```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm
```

```python
a = 1j/np.sqrt(2)
b = 1/np.sqrt(2)

state = np.array([0, 0.5, 0.5 + 0.7j, 0, 0.5, 0, 0, 0], dtype=complex)
state = state / np.linalg.norm(state)
```

```python
theta = 2 * np.arctan2(np.abs(a), np.abs(b))
RXX = expm(-1j*(-theta/2)*np.kron(np.array([[0, 1], [1, 0]]), np.array([[0, 1], [1, 0]])))
RYY = expm(-1j*(-theta/2)*np.kron(np.array([[0, -1j], [1j, 0]]), np.array([[0, -1j], [1j, 0]])))

U_x1 = np.kron(RXX, np.eye(2))
U_y1 = np.kron(RYY, np.eye(2))
U_1 = U_y1 @ U_x1

U_x2 = np.kron(np.eye(2), RXX)
U_y2 = np.kron(np.eye(2), RYY)
U_2 = U_y2 @ U_x2



new_state = U_2 @ U_1 @ state
n = np.pow(2, 10)

for i in range(n):
    new_state = U_2 @ U_1 @ new_state

# print("Original state:", state)
print("Transformed state:", new_state)
probabilities = np.abs(new_state)**2
print("\n\nProbabilities:", probabilities)
```
